```python
from SpectralClustering import SpectralClustering
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
model = SpectralClustering(clusters=3, sigma=0.3, neighbors=10)
kmeans = KMeans(n_clusters=3)
X, y = model.generateMultiMoons(num=1500, clusters=4, noise=0.05)
```

```python
ans = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=ans, s=30, cmap='viridis', edgecolor='k')
plt.title("KMeans")
plt.tight_layout()
plt.show()
```
